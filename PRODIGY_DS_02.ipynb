## Malaria Deaths Over Time in India using Stack plot

The stacked area plot illustrates the temporal trend of Malaria deaths in India over the years. Each layer of the stack corresponds to a specific year, showcasing the cumulative contribution of Malaria deaths.

```python
plt.figure(figsize=(6, 5))
India_df = India_Total_no_of_Deaths_df.groupby(['country', 'Year']).sum()
plt.stackplot(India_df.index.get_level_values('Year'), India_df['Malaria'], labels=India_df.index.get_level_values('Year'), color='blue', alpha=0.7)
plt.xlabel("Year")
plt.ylabel("Malaria Deaths")
plt.title(" Malaria Deaths Over Time in India")
plt.show()
```

## Chronic Kidney Disease Deaths Over Time in India using Area plot


This filled area plot visualizes the progression of Chronic Kidney Disease (CKD) deaths in India across different years. The dark violet-shaded areas represent the cumulative count of CKD deaths over time.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 6))
India_df = India_Total_no_of_Deaths_df.groupby(['country', 'Year']).sum()
plt.fill_between(India_df.index.get_level_values('Year'), 0, India_df['Chronic Kidney Disease'].cumsum(), color='darkviolet', alpha=0.7)
plt.xlabel("Year")
plt.ylabel("Chronic Kidney Disease")
plt.title("Chronic Kidney Disease Deaths Over Time in India")
plt.show()
```

# RANDOM UNDERSAMPLING

Undersampling is a technique used in machine learning to balance imbalanced datasets, where one class has significantly more samples than the other. In this case, the code is reducing the number of samples in the majority class to match the number of samples in the minority class.

```python
from imblearn.under_sampling import RandomUnderSampler
r=RandomUnderSampler(sampling_strategy=1)
x_res,y_res=r.fit_resample(x,y)

y_res.value_counts().plot.pie(autopct="%.2f").set_title("UNDERSAMPLING")
```

```python
y_res.value_counts()
```

# RANDOM OVERSAMPLING

Oversampling is a technique used in machine learning to balance imbalanced datasets, where one class has significantly fewer samples than the other. In this case, the code is increasing the number of samples in the minority class to match the number of samples in the majority class.

```python
from imblearn.over_sampling import RandomOverSampler
r=RandomOverSampler(sampling_strategy=1)
x_r,y_r=r.fit_resample(f,g)

y_r.value_counts().plot.pie(autopct="%.2f").set_title("OVERERSAMPLING")
```

# MODEL PREDICTION

## Splitting into training & testing data

```python
from sklearn.model_selection import train_test_split

x=df[['Meningitis',
       'Nutritional Deficiencies', 'Malaria', 'Drowning',
       'Interpersonal Violence', 'Maternal Disorders', 'HIV/AIDS',
       'Drug Use Disorders', 'Tuberculosis', 'Cardiovascular Diseases',
       'Lower Respiratory Infections', 'Neonatal Disorders',
       'Alcohol Use Disorders', 'Self-harm', 'Exposure to Forces of Nature',
       'Diarrheal Diseases', 'Environmental Heat and Cold Exposure',
       'Neoplasms', 'Conflict and Terrorism', 'Diabetes Mellitus',
       'Chronic Kidney Disease', 'Poisonings', 'Protein-Energy Malnutrition',
       'Road Injuries', 'Chronic Respiratory Diseases',
       'Cirrhosis and Other Chronic Liver Diseases', 'Digestive Diseases',
       'Fire, Heat, and Hot Substances', 'Acute Hepatitis', 'Death Count',
       'Total Deaths',]]
y=df['Death Count Category']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

## TESTING USING LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0,max_iter=500).fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report( y_test, y_pred))
```
